Let's compare Michigan to Ohio. Which state has the higher average outcome score across all years?

```python
michigan_score = df[df["STATE"] == "MICHIGAN"]
ohio_score = df[df["STATE"] == "OHIO"]

michigan_avg_score = michigan_score["AVG_MATH_4_SCORE"].mean()
ohio_avg_score = ohio_score["AVG_MATH_4_SCORE"].mean()

if michigan_avg_score > ohio_avg_score:
    print("Michigan")
else: 
    print ("Ohio")
```

Find the average for your outcome score across all states in 2019

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# create your model here
# model = Linear Regression
model = LinearRegression()
X = df[["AVG_READING_4_SCORE","AVG_MATH_8_SCORE","TOTAL_EXPENDITURE"]].dropna()
# y = The average math 4 score 
y = df.loc[X.index]["AVG_MATH_4_SCORE"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=0)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
print(model.intercept_)
model.score(X_test,y_test)
np.mean(model.predict(X_test)-y_test)
```
